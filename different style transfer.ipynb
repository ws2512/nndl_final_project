```python
import scipy.misc
```

```python
filenames = ["contents/Panda.jpg", "outputs/Panda_output.jpg",
             "outputs/Panda_output_2style.jpg", "outputs/Panda_output_preserve_color.jpg"]
images = []
for f in filenames:
    img = scipy.misc.imread(f)
    img = scipy.misc.imresize(img, (600, 800))
    images.append(img)
```

```python
r = 2
f, axarr = plt.subplots(r, r, figsize=(12,9))
for i in range(r):
    for j in range(r):
        img = images[r*i+j]
        axarr[i][j].imshow(img)
```
