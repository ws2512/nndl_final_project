```python
import scipy.misc

filenames = ["contents/Paris_FR.jpg", "outputs/Paris_FR_output_adam_lr_1e-2.jpg",
             "outputs/Paris_FR_output_adam_lr_1e-1.jpg", "outputs/Paris_FR_output_adam_lr_1.0.jpg",
             "outputs/Paris_FR_output_adam_lr_2.0.jpg", "outputs/Paris_FR_output_adam_lr_10.0.jpg"]
images = []
for f in filenames:
    img = scipy.misc.imread(f)
    img = scipy.misc.imresize(img, (600, 800))
    images.append(img)
```

```python
r1 = 3 
r2 = 2
f, axarr = plt.subplots(r1, r2, figsize=(10,12))
for i in range(r1):
    for j in range(r2):
        img = images[r2*i+j]
        axarr[i][j].imshow(img)
```
