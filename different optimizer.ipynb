```python
import tensorflow as tf
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from Artistic_Style import ArtisticStyle
art = ArtisticStyle()
```

```python
%%time
tf.reset_default_graph()
costs_adam = art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_adam", alpha=5, beta=1000, optimizer='Adam', 
                 learning_rate=0.2, iterations=1000, verbose=True)
```

```python
%%time
tf.reset_default_graph()
costs_adagrad = art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_adagrad", alpha=5, beta=1000, optimizer='Adagrad', 
                 learning_rate=0.2, iterations=1000, verbose=True)
```

```python
%%time
tf.reset_default_graph()
costs_adadelta = art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_adadelta", alpha=5, beta=1000, optimizer='Adadelta', 
                 learning_rate=0.2, iterations=1000, verbose=True)
```

```python
%%time
tf.reset_default_graph()
costs_rmsprop = art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_rmsprop", alpha=5, beta=1000, optimizer='Rmsprop', 
                 learning_rate=0.2, iterations=1000, verbose=True)
```

```python
%%time
tf.reset_default_graph()
costs_sgd = art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_sgd", alpha=5, beta=1000, optimizer='Sgd', 
                 learning_rate=0.2, iterations=1000, verbose=True)
```

```python
%%time
tf.reset_default_graph()
art.training(content='NewYork', styles=['Sunrise'], 
                 style_weights=[1], output="NewYork_lbfgsb", alpha=5, beta=1000, optimizer='Lbfgsb', 
                 learning_rate=0.2, iterations=1000, verbose=False)
```

```python
plt.plot(costs_adam, label="adam")
plt.plot(costs_adagrad, label="adagrad")
plt.plot(costs_adadelta, label="adadelta")
plt.plot(costs_rmsprop, label="rmsprop")
plt.plot(costs_sgd, label="sgd")
plt.ylim([0, 5e10])
plt.legend(bbox_to_anchor=(1, 1.05))
plt.show()
```

```python
import scipy.misc
filenames = ["contents/NewYork.jpg", "outputs/NewYork_adam.jpg",
             "outputs/NewYork_adagrad.jpg", "outputs/NewYork_adadelta.jpg",
             "outputs/NewYork_rmsprop.jpg", "outputs/NewYork_sgd.jpg",
             "outputs/NewYork_lbfgsb.jpg"]
images = []
for f in filenames:
    img = scipy.misc.imread(f)
    img = scipy.misc.imresize(img, (600, 800))
    images.append(img)
```
